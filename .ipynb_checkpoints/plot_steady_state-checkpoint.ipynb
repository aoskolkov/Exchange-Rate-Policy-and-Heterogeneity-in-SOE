Options for calculating the steady state:

* baseline productivity differences: productivity in non-tradables is 0.6 of that in tradables
* equal productivities
* productivities switched: productivity in tradables is 0.6 of that in non-tradables

Choose *productivity_gap* from mutually exclusive "**baseline**", "**equal**", "**switched**" (uncomment the relevant one):

```julia
productivity_gap = "baseline"
#productivity_gap = "equal"
#productivity_gap = "switched"
```

```julia
# loading parameters
if productivity_gap == "baseline"
    @load "external_parameters.jld2"
    @load "steady_state_output.jld2"
elseif productivity_gap == "equal"
    @load "external_parameters_equal.jld2"
    @load "steady_state_output_equal.jld2"
elseif productivity_gap == "switched"
    @load "external_parameters_switched.jld2"
    @load "steady_state_output_switched.jld2"
end;
```

```julia
# compute and plot MPC for the two sectors
par_grid_T, par_grid_N = deepcopy(par_grid), deepcopy(par_grid)
par_grid_N.productivity, par_grid_N.labor = prod_N*par_grid_N.productivity, prod_N*par_grid_N.labor
par_grid_T.productivity, par_grid_T.labor = prod_T*par_grid_T.productivity, prod_T*par_grid_T.labor

@time paral_mpc = pmap(compute_mpc,[c_quarter_T,c_quarter_N],[cons_T,cons_N],[value_T,value_N],[distr_T,distr_N],[par,par],[par_grid_T,par_grid_N])
MPC_T, MPC_bydecile_T = paral_mpc[1]
MPC_N, MPC_bydecile_N = paral_mpc[2]
print("\n")

a = sum((par.ζ_T*distr_T+par.ζ_N*distr_N).*par_grid.assets)
distrN_a = sum(reshape(distr_N,par_grid.N,par_grid.K),dims=2)[:]
distrT_a = sum(reshape(distr_T,par_grid.N,par_grid.K),dims=2)[:]
distr_a = par.ζ_T*distrT_a+par.ζ_N*distrN_a

plot_distr = plot(par_grid.q_grid/a,distr_a/maximum(distr_a)*0.75,color=:grey,label="")
plot!(par_grid.q_grid/a,zeros(par_grid.N),fillrange=distr_a/maximum(distr_a)*0.75,linecolor=:white,color=:blue,alpha=0.3,linewidth=3,label=L"\mathrm{density}",xlim=[par_grid.a_bar/a,5],legend=:topright,xlabel=L"a",title=L"\mathrm{(a)\ MPC\ and\ asset\ distribution}")
plot!(par_grid.q_grid[2:end]/a,par.ζ_T*MPC_T+par.ζ_N*MPC_N,color=:red,linewidth=3,label=L"\mathrm{MPC}",linestyle=:dash,ylim=[-0.01,0.4])
mpc_plot = plot(par.ζ_T*MPC_bydecile_T+par.ζ_N*MPC_bydecile_N,st=:scatter,color=:purple,markersize=6,ma=0.75,label=L"\mathrm{model}",ylim=[-0.01,0.6],title=L"\mathrm{(c)\ MPC\ across\ income\ deciles}")
plot!(MPC_target,st=:scatter,color=:black,markersize=6,ma=0.75,markershape=:diamond,label=L"\mathrm{data}")
plot_gap = plot(par_grid.q_grid/a,cumsum(distrN_a),xlim=[par_grid.a_bar/a,9],lw=2,color=:firebrick,label=L"\mathrm{N}",legend=:bottomright,xlabel=L"a",title=L"\mathrm{(b)\ CDF\ in\ the\ two\ sectors}")
plot!(par_grid.q_grid/a,cumsum(distrT_a),color=:navy,ls=:dash,lw=2,label=L"\mathrm{T}")
plot_all = plot(plot_distr,plot_gap,mpc_plot,layout=@layout[a a a],size=(900,300),titlefontsize=12,legendfontsize=12,tickfontsize=9,bottom_margin=0.35Plots.cm)
```
