```julia
using Compat, Random, Distributions, Plots, LinearAlgebra, Statistics, LaTeXStrings, SparseArrays, CPUTime, JLD2
using SpecialFunctions, GSL, SuiteSparse, Distributed, Roots, NBInclude
gr(fmt=:png)
e = 2.7182818284
Random.seed!(1);
```

```julia
addprocs(2)
@everywhere using Compat, Random, Distributions, Plots, LinearAlgebra, Statistics, LaTeXStrings, SparseArrays, CPUTime, JLD2, SpecialFunctions, GSL, SuiteSparse, Distributed
@everywhere using SharedArrays, ParallelDataTransfer, Roots, NBInclude
@everywhere @nbinclude("setup.ipynb")
@everywhere using .setup, .functions
procs()
```

```julia
# set the time grid
T = 600
dt = T/1200
T_bar = 24
time_grid = collect(0:dt:T)
L = length(time_grid)
t_grid = dt, L;
```

```julia
# compute steady-state statistics
@load "steady_state_output.jld2"
@load "external_parameters.jld2"
par_grid_N, par_grid_T = deepcopy(par_grid), deepcopy(par_grid)
par_grid_N.labor = par_grid.labor * prod_N
par_grid_N.productivity = par_grid.productivity * prod_N
par_grid_T.labor = par_grid.labor * prod_T
par_grid_T.productivity = par_grid.productivity * prod_T
aN = distr_N'*par_grid.assets
aT = distr_T'*par_grid.assets
a = par.ζ_T*aT + par.ζ_N*aN

@time paral = pmap(q_response,[value_N,value_T],[distr_N,distr_T],(1-par.τ_w)*[ones(L),ones(L)],[par.r_d*ones(L),par.r_d*ones(L)],[zeros(L),zeros(L)],[true,true],[par,par],[par_grid_N,par_grid_T],[t_grid,t_grid])
 
Cq_N, Lq_N, v_N = paral[1]
Cq_T, Lq_T, v_T = paral[2];
```

```julia
# compute the response decomposition for float
@load "transition_results/float.jld2"
res_float = res_this

@time paral = pmap(q_response,[value_N,value_T],[distr_N,distr_T],(1-par.τ_w)*[res_float.wN_sequence,res_float.wT_sequence],[res_float.rd_sequence,res_float.rd_sequence],[zeros(L),zeros(L)],[true,true],[par,par],[par_grid_N,par_grid_T],[t_grid,t_grid])
Call_N_float, Lall_N_float, Wall_N_float = paral[1]
Call_T_float, Lall_T_float, Wall_T_float = paral[2]

@time paral = pmap(q_response,[value_N,value_T],[distr_N,distr_T],(1-par.τ_w)*[res_float.wN_sequence,res_float.wT_sequence],[par.r_d*ones(L),par.r_d*ones(L)],[zeros(L),zeros(L)],[true,true],[par,par],[par_grid_N,par_grid_T],[t_grid,t_grid])
Cwage_N_float, Lwage_N_float, Wwage_N_float = paral[1]
Cwage_T_float, Lwage_T_float, Wwage_T_float = paral[2]

@time paral = pmap(q_response,[value_N,value_T],[distr_N,distr_T],(1-par.τ_w)*[ones(L),ones(L)],[res_float.rd_sequence,res_float.rd_sequence],[zeros(L),zeros(L)],[true,true],[par,par],[par_grid_N,par_grid_T],[t_grid,t_grid])
Crate_N_float, Lrate_N_float, Wrate_N_float = paral[1]
Crate_T_float, Lrate_T_float, Wrate_T_float = paral[2]

full_float_T = reshape(100*log.(Call_T_float./Cq_T),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
wage_float_T = reshape(100*log.(Cwage_T_float./Cq_T),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
rate_float_T = reshape(100*log.(Crate_T_float./Cq_T),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K

full_float_N = reshape(100*log.(Call_N_float./Cq_N),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
wage_float_N = reshape(100*log.(Cwage_N_float./Cq_N),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
rate_float_N = reshape(100*log.(Crate_N_float./Cq_N),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K

# compute the response decomposition for peg
@load "transition_results/peg.jld2"
res_peg = res_this

@time paral = pmap(q_response,[value_N,value_T],[distr_N,distr_T],(1-par.τ_w)*[res_peg.wN_sequence,res_peg.wT_sequence],[res_peg.rd_sequence,res_peg.rd_sequence],[zeros(L),zeros(L)],[true,true],[par,par],[par_grid_N,par_grid_T],[t_grid,t_grid])
Call_N_peg, Lall_N_peg, Wall_N_peg = paral[1]
Call_T_peg, Lall_T_peg, Wall_T_peg = paral[2]

@time paral = pmap(q_response,[value_N,value_T],[distr_N,distr_T],(1-par.τ_w)*[res_peg.wN_sequence,res_peg.wT_sequence],[par.r_d*ones(L),par.r_d*ones(L)],[zeros(L),zeros(L)],[true,true],[par,par],[par_grid_N,par_grid_T],[t_grid,t_grid])
Cwage_N_peg, Lwage_N_peg, Wwage_N_peg = paral[1]
Cwage_T_peg, Lwage_T_peg, Wwage_T_peg = paral[2]

@time paral = pmap(q_response,[value_N,value_T],[distr_N,distr_T],(1-par.τ_w)*[ones(L),ones(L)],[res_peg.rd_sequence,res_peg.rd_sequence],[zeros(L),zeros(L)],[true,true],[par,par],[par_grid_N,par_grid_T],[t_grid,t_grid])
Crate_N_peg, Lrate_N_peg, Wrate_N_peg = paral[1]
Crate_T_peg, Lrate_T_peg, Wrate_T_peg = paral[2]

full_peg_T = reshape(100*log.(Call_T_peg./Cq_T),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
wage_peg_T = reshape(100*log.(Cwage_T_peg./Cq_T),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
rate_peg_T = reshape(100*log.(Crate_T_peg./Cq_T),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K

full_peg_N = reshape(100*log.(Call_N_peg./Cq_N),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
wage_peg_N = reshape(100*log.(Cwage_N_peg./Cq_N),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
rate_peg_N = reshape(100*log.(Crate_N_peg./Cq_N),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K;

# plot the decomposition
plot_CT_float = plot(par_grid.q_grid/a,full_float_T,xlim=[par_grid.a_bar/a,5.0],color=:black,ls=:dash,lw=3,ylim=[-0.95,0.25],label="",title=L"\mathrm{(a)\ T\ sector,\ float}")
plot!(par_grid.q_grid/a,zeros(par_grid.N),fillrange=wage_float_T,color=:deepskyblue,alpha=0.33,label=L"\mathrm{wage\ effect}")
plot!(par_grid.q_grid/a,zeros(par_grid.N),fillrange=rate_float_T,color=:grey,alpha=0.5,grid=:false,label=L"\mathrm{interest\ rate\ effect}")
plot!(par_grid.q_grid/a,full_float_T,color=:black,ls=:dash,lw=3,label="")

plot_CN_float = plot(par_grid.q_grid/a,full_float_N,xlim=[par_grid.a_bar/a,5.0],color=:black,ls=:dash,lw=3,ylim=[-0.95,0.25],label="",title=L"\mathrm{(b)\ N\ sector,\ float}")
plot!(par_grid.q_grid/a,rate_float_N,fillrange=rate_float_N+wage_float_N,color=:deepskyblue,alpha=0.33,label=L"\mathrm{wage\ effect}")
plot!(par_grid.q_grid/a,zeros(par_grid.N),fillrange=rate_float_N,color=:grey,alpha=0.5,grid=:false,label=L"\mathrm{interest\ rate\ effect}")
plot!(par_grid.q_grid/a,full_float_N,color=:black,ls=:dash,lw=3,label="")

lab_N = reshape(100*log.(Lall_N_float./Lq_N),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
lab_T = reshape(100*log.(Lall_T_float./Lq_T),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
plot_L_float = plot(par_grid.q_grid/a,lab_N,xlim=[par_grid.a_bar/a,5.0],color=:black,lw=3,label=L"\mathrm{N}",title=L"\mathrm{(c)\ labor\ response,\ float}",ylim=[-0.4,0.4])
plot!(par_grid.q_grid/a,lab_T,color=:grey,ls=:dash,lw=3,label=L"\mathrm{T}")

plot_CT_peg = plot(par_grid.q_grid/a,full_peg_T,xlim=[par_grid.a_bar/a,5.0],color=:black,ls=:dash,lw=3,ylim=[-0.95,0.25],label="",title=L"\mathrm{(d)\ T\ sector,\ peg}")
plot!(par_grid.q_grid/a,zeros(par_grid.N),fillrange=wage_peg_T,color=:red,alpha=0.25,label=L"\mathrm{wage\ effect}")
plot!(par_grid.q_grid/a,zeros(par_grid.N),fillrange=rate_peg_T,color=:grey,alpha=0.5,grid=:false,label=L"\mathrm{interest\ rate\ effect}")
plot!(par_grid.q_grid/a,full_peg_T,color=:black,ls=:dash,lw=3,label="")

plot_CN_peg = plot(par_grid.q_grid/a,full_peg_N,xlim=[par_grid.a_bar/a,5.0],color=:black,ls=:dash,lw=3,ylim=[-0.95,0.25],label="",title=L"\mathrm{(e)\ N\ sector,\ peg}")
plot!(par_grid.q_grid/a,rate_peg_N,fillrange=rate_peg_N+wage_peg_N,color=:red,alpha=0.25,label=L"\mathrm{wage\ effect}")
plot!(par_grid.q_grid/a,zeros(par_grid.N),fillrange=rate_peg_N,color=:grey,alpha=0.5,grid=:false,label=L"\mathrm{interest\ rate\ effect}")
plot!(par_grid.q_grid/a,full_peg_N,color=:black,ls=:dash,lw=3,label="")

lab_N = reshape(100*log.(Lall_N_peg./Lq_N),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
lab_T = reshape(100*log.(Lall_T_peg./Lq_T),par_grid.N,par_grid.K)*ones(par_grid.K)/par_grid.K
plot_L_peg = plot(par_grid.q_grid/a,lab_N,xlim=[par_grid.a_bar/aN,5.0],color=:black,lw=3,label=L"\mathrm{N}",title=L"\mathrm{(f)\ labor\ response, peg}",ylim=[-0.4,0.4])
plot!(par_grid.q_grid/a,lab_T,color=:grey,ls=:dash,lw=3,label=L"\mathrm{T}")

l = @layout[a a a; a a a]
plot(plot_CT_float,plot_CN_float,plot_L_float,plot_CT_peg,plot_CN_peg,plot_L_peg,layout=l,size=(900,600),legendfontsize=12,titlefontsize=12,legend=:bottomright,xlabel=L"a")
```
