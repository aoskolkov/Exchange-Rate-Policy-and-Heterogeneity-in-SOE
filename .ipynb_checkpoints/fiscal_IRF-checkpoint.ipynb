```julia
using Compat, Random, Distributions, Plots, LinearAlgebra, Statistics, LaTeXStrings, SparseArrays, CPUTime, JLD2
using SpecialFunctions, GSL, SuiteSparse, Distributed, Roots, NBInclude
gr(fmt=:png)
e = 2.7182818284
@nbinclude("setup.ipynb")
Random.seed!(1);
```

```julia
# set the time grid
T = 600
dt = T/1200
T_bar = 24
time_grid = collect(0:dt:T)
L = length(time_grid)
 = 0.02/12*exp.(-time_grid/12*1.15)

# set up matrices that convert monthly responses to quarterly
Lq = Int64((L-1)/8)
A = Diagonal(ones(Lq))
b = ones(8)'/8
Q = hcat(kron(A,b),zeros(Lq))
Tq_bar = 11
tq_grid = time_grid[9:8:end]/4

# load parameters
@load "steady_state_output.jld2"

# plot IRF
@load "transition_results/float.jld2"
aggr_gov = res_this.qmain_sequence
tax_gov = par._w*(res_this.wT_sequence.*res_this.laborT_sequence*par._T+res_this.wN_sequence.*res_this.qN_sequence)
prof_gov = (ones(L) - res_this.marg_sequence).*res_this.qmain_sequence
debt_gov = - (res_this.rd_sequence-ones(L)*par.r_d)*debt
gov_seq = res_this.qmain_sequence - (par._T * res_this.cT_sequence + par._N * res_this.cN_sequence)

@load "transition_results/float_prop.jld2"
aggr_prop = res_this.qmain_sequence

@load "transition_results/float_flat.jld2"
aggr_flat = res_this.qmain_sequence

plot_aggr_float = plot(tq_grid,100*(Q*aggr_gov/aggr_gov[end]-ones(Lq)),xlim=[1,Tq_bar],color=:grey,lw=2,label=L"\mathrm{government}",legend=:bottomright,ylim=[-0.85,0.15],title=L"\mathrm{(a)\ agg.\ demand,\ float}")
plot!(tq_grid,100*(Q*aggr_prop/aggr_prop[end]-ones(Lq)),xlim=[1,Tq_bar],color=:black,lw=2,ls=:dash,label=L"\mathrm{taxes\ (neutral)}")
plot!(tq_grid,100*(Q*aggr_flat/aggr_flat[end]-ones(Lq)),xlim=[1,Tq_bar],color=:black,lw=2,ls=:dot,label=L"\mathrm{taxes\ (flat)}")
plot_decomp_float = plot(tq_grid,100*(Q*tax_gov/gov_seq[end]-ones(Lq)*tax_gov[end]/gov_seq[end]),xlim=[1,Tq_bar],color=:blue,lw=2,label=L"\mathrm{taxes}",title=L"\mathrm{(b)\ gov.\ surplus,\ float}",ylim=[-3.5,1.5])
plot!(tq_grid,100*(Q*debt_gov/gov_seq[end]-ones(Lq)*debt_gov[end]/gov_seq[end]),color=:red,ls=:dash,lw=2,label=L"\mathrm{debt\ servicing}")
plot!(tq_grid,100*(Q*prof_gov/gov_seq[end]-ones(Lq)*prof_gov[end]/gov_seq[end]),color=:black,ls=:dot,lw=2,label=L"\mathrm{profits}")

@load "transition_results/peg.jld2"
aggr_gov = res_this.qmain_sequence
tax_gov = par._w*(res_this.wT_sequence.*res_this.laborT_sequence*par._T+res_this.wN_sequence.*res_this.qN_sequence)
prof_gov = (ones(L) - res_this.marg_sequence).*res_this.qmain_sequence
debt_gov = - (res_this.rd_sequence-ones(L)*par.r_d)*debt
gov_seq = res_this.qmain_sequence - (par._T * res_this.cT_sequence + par._N * res_this.cN_sequence)

@load "transition_results/peg_prop.jld2"
aggr_prop = res_this.qmain_sequence

@load "transition_results/peg_flat.jld2"
aggr_flat = res_this.qmain_sequence

plot_aggr_peg = plot(tq_grid,100*(Q*aggr_gov/aggr_gov[end]-ones(Lq)),xlim=[1,Tq_bar],color=:grey,lw=2,label=L"\mathrm{government}",legend=:bottomright,ylim=[-0.85,0.15],title=L"\mathrm{(c)\ agg.\ demand,\ peg}")
plot!(tq_grid,100*(Q*aggr_prop/aggr_prop[end]-ones(Lq)),xlim=[1,Tq_bar],color=:black,lw=2,ls=:dash,label=L"\mathrm{taxes\ (neutral)}")
plot!(tq_grid,100*(Q*aggr_flat/aggr_flat[end]-ones(Lq)),xlim=[1,Tq_bar],color=:black,lw=2,ls=:dot,label=L"\mathrm{taxes\ (flat)}")
plot_decomp_peg = plot(tq_grid,100*(Q*tax_gov/gov_seq[end]-ones(Lq)*tax_gov[end]/gov_seq[end]),xlim=[1,Tq_bar],color=:blue,lw=2,label=L"\mathrm{taxes}",title=L"\mathrm{(d)\ gov.\ surplus,\ peg}",ylim=[-3.5,1.5])
plot!(tq_grid,100*(Q*debt_gov/gov_seq[end]-ones(Lq)*debt_gov[end]/gov_seq[end]),color=:red,ls=:dash,lw=2,label=L"\mathrm{debt\ servicing}")
plot!(tq_grid,100*(Q*prof_gov/gov_seq[end]-ones(Lq)*prof_gov[end]/gov_seq[end]),color=:black,ls=:dot,lw=2,label=L"\mathrm{profits}")

l = @layout [a a a a]
plot(plot_aggr_float,plot_decomp_float,plot_aggr_peg,plot_decomp_peg,layout=l,size=(900,250),legend=:bottomright,legendfontsize=12,tickfontsize=9,titlefontsize=12)
```
